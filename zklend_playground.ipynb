## How I assume things work

1) [`Deposit`] User deposits to Zklend. [`Withdrawal`] User can also withdraw from Zklend.
2) [`CollateralEnabled`] User marks the deposit (or part of it) as collateral. [`CollateralDisabled`] User can also mark the collateral (or part of it) as non-collateral, the amount then stays at Zklend as deposit.
3) [`Borrowing`] User borrows from Zklend.
4) [`Repayment`] User repays his loan.
5) [`Liquidation`] Someone liquidates (part of?) the user's loan.
